```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

```

```python
# Loading Data
data = pd.read_csv( 'datasets/kc_house_data.csv' )
#Formato date
data['date'] = pd.to_datetime( data['date'], format='%Y-%m-%d' )
```

Faça um gráfico de barras que represente a soma dos preços pelo número
de quartos.

```python
plt.figure( figsize=(10,6))
df = data[['price', 'bedrooms']].groupby( 'bedrooms').sum().reset_index()
sns.barplot( palette='BuGn_d', x='bedrooms', y='price', data=df );

```

Faça um gráfico de linhas que represente a média dos preços pelo ano
construção dos imóveis.


```python
#estratégia; selecionar as colunas:'prices' e 'yr_built', guardar em uma variavel df, e realizar a media, após isso, selecionar uma função de seaborn que crie um gráfico de linhas
df = data[['price', 'yr_built']].groupby( 'yr_built' ).mean().reset_index()
sns.lineplot( x='yr_built', y='price', data=df );

```

 Faça um gráfico de barras que represente a média dos preços pelo tipo
dos dormitórios.


```python
data['dormitory_type'] = data['bedrooms'].apply( lambda x: 'studio' if x == 1
                                                else 'apartament' if x == 2 else 'house' )

plt.figure( figsize=(10,6))
df = data[['price', 'dormitory_type']].groupby( 'dormitory_type' ).mean().reset_index()
sns.barplot(  palette='pink', x='dormitory_type', y='price', data=df );
```

Faça um gráfico de linha que mostre a evolução da média dos preços pelo
ano da reforma dos imóveis, a partir do ano de 1930.

```python
plt.figure( figsize=(24,12))
df = data.loc[data['yr_renovated'] > 1930, ['price', 'yr_renovated']].groupby('yr_renovated' ).mean().reset_index()
sns.lineplot( x='yr_renovated', y='price', data=df );
                          
```

Faça um tabela que mostre a média dos preços por ano de construção e
tipo de dormitórios dos imóveis.


```python
df = data[['price', 'yr_built', 'dormitory_type']].groupby( ['yr_built','dormitory_type'] ).mean().reset_index()
df.head()
```

7. Crie um Dashboard com os gráficos das questões 02, 03, 04 ( Dashboard:
1 Linha e 2 colunas )

```python
from matplotlib import gridspec
fig = plt.figure( figsize=(24, 12) )
specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )
ax1 = fig.add_subplot( specs[0, :] ) # First Row
ax2 = fig.add_subplot( specs[1, 0] ) # First Row First Column
ax3 = fig.add_subplot( specs[1, 1] ) # Second Row First Column
# Questão 2
df = data[['price', 'bedrooms']].groupby( 'bedrooms' ).sum().reset_index()
sns.barplot(  x='bedrooms', y='price', data=df );
ax1.bar( df['bedrooms'], df['price'] )
# Questão 3
df = data[['price', 'yr_built']].groupby( 'yr_built' ).mean().reset_index()
ax2.plot( df['yr_built'], df['price'] );
# Questão 4
df = data[['price', 'dormitory_type']].groupby( 'dormitory_type' ).mean().reset_index()
ax3.bar( df['dormitory_type'], df['price'] );
```

8. Crie um Dashboard com os gráficos das perguntas 02, 04 ( Dashboard:
2 colunas )

```python
fig = plt.figure( figsize=(24, 12) )
specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )
ax1 = fig.add_subplot( specs[0, 0] ) # First Row First Column
ax2 = fig.add_subplot( specs[0, 1] ) # Second Row First Column
# Questão 2
df = data[['price', 'bedrooms']].groupby( 'bedrooms' ).sum().reset_index()
sns.barplot( x='bedrooms', y='price', data=df );
ax1.bar( df['bedrooms'], df['price'] )
# Questão 4
df = data[['price', 'dormitory_type']].groupby( 'dormitory_type' ).mean().reset_index()
ax2.bar( df['dormitory_type'], df['price'] );
```

Crie um Dashboard com os gráficos das perguntas 03, 05 ( Dashboard:
2 Linhas )

```python
fig = plt.figure( figsize=(24, 12) )
specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )
ax1 = fig.add_subplot( specs[0, 0] ) # First Row First Column
ax2 = fig.add_subplot( specs[1, 0] ) # Second Row First Column
# Frist Graph
df = data[['price', 'yr_built']].groupby( 'yr_built' ).mean().reset_index()
ax1.plot( df['yr_built'], df['price'] );
# Second Graph
df = data.loc[data['yr_renovated'] > 1930, ['price', 'yr_renovated']].groupby('yr_renovated' ).mean().reset_index()
ax2.plot( df['yr_renovated'], df['price'] );
```

Faça um gráfico com o tamanho dos pontos sendo igual ao tamanho
da sala de estar


```python
houses = data[['id', 'lat', 'long', 'price', 'sqft_living']].copy()
# define level of prices
houses['level'] = houses['price'].apply( lambda x: 0 if x <= 321950 else
1 if (x > 321950 ) & ( x <= 450000 ) else
2 if (x > 450000 ) & ( x <= 645000 ) else 3 )
# convert level to astype
houses['level'] = houses['level'].astype( int )
fig = px.scatter_mapbox( houses,
lat="lat",
lon="long",
color='level',
size="sqft_living",
color_continuous_scale=px.colors.cyclical.IceFire,
size_max=15,
zoom=10)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(height=600, margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
